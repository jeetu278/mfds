```python
import pandas as pd
import matplotlib.pyplot as plt
Url='/content/Customer Purchase.csv'
df=pd.read_csv(Url)
df.dtypes
```

```python
df.head()
```

```python
df
```

```python
mean_value=df['Age'].mean()
print(f"mean value:- {mean_value}")
```

```python
median_value=df['Age'].median()
print(f"median value:- {median_value}")
```

```python
mode_value=df['Age'].mode()
print(f"mode value:- {mode_value}")
```

```python
var_value=df['Age'].var()
print(f"variance value:- {var_value}")
```

```python
std_value=df['Age'].std()
print(f"standard deviation value:- {std_value}")
```

```python
skew_value=df['Age'].skew()
print(f"skewness value:- {skew_value}")
```

```python
kur_value=df['Age'].kurt()
print(f"kurtosis value:- {kur_value}")
```

```python
print(df.columns)
```

```python
average=df.groupby('Customer ID')['Age'].mean()
```

```python
average.plot(kind='bar', color='skyblue')
plt.xlabel('Customer ID')
plt.ylabel('Age')
plt.show()
```
